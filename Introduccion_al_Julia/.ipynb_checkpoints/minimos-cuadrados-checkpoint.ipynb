# Mínimos cuadrados

** Problema de Mínimos Cuadrados (PMC) :** 
> Dada $A\in\mathbb{C}^{m\times n}$, ($m>n$), $b\in \mathbb{C}^m$, encontrar $x\in\mathbb{C}^n$ tal que $\|Ax-b \|$ es mínimo.


## Aproximación por polinomios
### Interpolación vs Mínimos Cuadrados

Considerando los siguientes datos $\{(x_i,y_i):\; i=1\ldots m\}$, encontrar un polinomio que aproxime los datos.

#### Interpolación 


```julia
using PyPlot #llamamos paquete para graficar
```

```julia
#definimos m y los datos
m=10;
x=rand(m);
y=abs(10*randn(m));
```

```julia
# Matriz de Vandermonde
function vandermonde(x,columnas=length(x))
    m=length(x)
    A=zeros(m,columnas);
    for k=1:columnas
    A[:,k]=x.^(k-1)
    end
    return A
end
```

```julia
A=vandermonde([1,2,3,4],2)
```

```julia
A=vandermonde(x); #matriz de vandermonde 
```

```julia
# Hallamos coeficientes de interpolación
cint=A\y; #resolvemos sistema lineal Ac=y 
```

```julia
# Para definir polinomio interpolador
# definimos la función evaluación de un polinomio
function poly(z,coef=cint)
    a=0
    grado=length(coef)
    for k=0:grado-1
        a=a+coef[k+1]*z^k 
    end
    return a
end
```

```julia
# Verificamos que polinomio interpola correctamente
    print(poly(x[10])-y[10])
```

```julia
z=linspace(minabs(x),maxabs(x),1000)#partición del dominio
plot(z,[poly(z[k]) for k in 1:length(z)];) # gráfica de polinomio
plot(x,y,"*") # gráfica de datos
```

#### Mínimos cuadrados

Busquemos el polinomio de grado $n$ (con $m>n$) que sea solucion de mínimo cuadrados para los datos dados.

```julia
n=7; A=vandermonde(x,n);
```

```julia
Q,R=qr(A); # descomposición QR de la matriz $A$
```

```julia
cmin=R\Q'y; # hallamos coeficiéntes de la solución PMC
```

```julia
plot(z,[poly(z[k],cmin) for k in 1:length(z)],"red") # gráfica de polinomio
plot(x,y,"*") # gráfica de datos
```

### Comparación de interpolación vs mínimos cuadrados

```julia
plot(z,[poly(z[k],cmin) for k in 1:length(z)],"red") # gráfica de polinomio
plot(z,[poly(z[k]) for k in 1:length(z)];) # gráfica de polinomio
plot(x,y,"*") # gráfica de datos
```

#### Simulación interactiva

```julia
using Interact
```

```julia
#Fijamos datos
x=rand(10);
y=abs(10*randn(10));
A=vandermonde(x);
cint=A\y;
```

```julia
function doplot(n)
     B=vandermonde(x,n);
    Q,R=qr(B); # descomposición QR de la mat
    cmin=R\Q'y;
    z=linspace(minabs(x),maxabs(x),100)
    plot(z,[poly(z[k],cmin) for k in 1:length(z)],"red") # gráfica de polinomio
    plot(z,[poly(z[k],cint) for k in 1:length(z)];) # gráfica de polinomio
    plot(x,y,"*") # gráfica de datos
    #return f
end
```

```julia
f=figure()
#f=figure(4,4)
@manipulate for n in 1:length(x); withfig(f) do 
        doplot(n)
    end
end
```
