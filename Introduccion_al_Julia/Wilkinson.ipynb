# Polinomio de Wilkinson

```julia
using Polynomials
```

```julia
using PyPlot
```

```julia
c=collect([1:20])
c=int128(c)
```

```julia
Wil=poly(c)
```

```julia
coef=coeffs(Wil)
```

```julia
function Wileval(x)
    prod([x-k for k in 1:20])
    end
```

```julia
Wileval(18.00000000001)
```

```julia
Wileval(17.99999999999)
```

```julia
coefper=coef;
coefper[15]=coefper[15]+10.0^(-10)
```

```julia
Wilper=Poly(coefper)
```

```julia
r=roots(Wilper)
```

```julia
function Wilpereval(x)
    prod([x-r[k] for k in 1:20])
    end
```

```julia
z=linspace(15,18.5,1000);
```

```julia
plot(z,[Wileval(k) for k in z])
plot(z,[Wilpereval(k) for k in z])
plot(z,[0 for k in z])
```
