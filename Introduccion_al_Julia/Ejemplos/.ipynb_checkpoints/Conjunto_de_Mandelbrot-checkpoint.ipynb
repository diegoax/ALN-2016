# Ejemplos: conjuntos de Julia y conjunto de Mandelbrot

# Conjunto de Mandelbrot

El conjunto de Mandelbrot $\mathcal M\subset\mathbb C$ se define como el conjunto de valores $c\in \mathbb{C}$ tales que la iteración de $c$ por la función $f(z)=z^2+c$ permanece acotada.

Es fácil ver que si $|z|>2$, entonces $\lim_{n}f^n(z)=\infty$, donde $\infty\in\overline{\mathbb{C}}$ (plano complejo ampliado). Un resultado de *dinámica compleja* es que si $z\in\mathcal{M}^c$, entonces existe $k\in\mathbb N$ tal que $|f^k(z)|>2$. De esta manera podemos describir un *algoritmo* que "aproxime" $\mathcal M^c$.

```julia
using PyPlot #llamamos paquete para graficar
```

El *output* de la siguiente función, con *input* $z$, es el primer $k\in [1,\ldots,N]$ tal que $|f^k(z)|>2$. De no haberlo, el output es $N$.

```julia
function f(z,c; N=300)
    for n = 1:N
        if abs2(z) > 2 # abs2 = |2|
            return n-1
        end
        z = z*z + c
    end
    return N
end
```

Definimos el arreglo __m__ como output de la función mandelbrot en puntos de una grilla. 
La variable **@time** nos dá información sobre el tiempo y memoria del cálculo. 

```julia
@time m = [ f(complex(r,i),complex(r,i)) for i=1:-.002:-1, r=-1.5:.002:1.5 ];
```

```julia
get_cmap("RdGy") # mapa de colores 
```

```julia
imshow(m, cmap="RdGy", extent=[-1.5,1.5,-1,1]) # mostrar imagen
```

## Ejemplo de conjunto de Julia

```julia
@time m = [ f(complex(r,i),complex(0.4,0.5)) for i=1:-.002:-1, r=-1.5:.002:1.5 ];
```

```julia
imshow(m, cmap="RdGy", extent=[-1.5,1.5,-1,1])
```
