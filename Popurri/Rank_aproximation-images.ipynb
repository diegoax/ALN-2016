```julia
using PyPlot
```

```julia
img=imread("mariobros.png"); # carga imagen
```

```julia
imshow(img) # muestra imagen
```

```julia
R=img[:,:,1]; A=1-R; # alterna colores blancos y negros
```

```julia
imshow(A;cmap=:Greys) # muestra imagen en escala de grises
```

```julia
U,,V=svd(A) # descomposicin en valores singulares de la matriz A
r=40  # rango de la aproximacin
imshow(U[:,1:r]*diagm([1:r])*V[:,1:r]';cmap=:Greys) # muestra imagen aproximada.
```
