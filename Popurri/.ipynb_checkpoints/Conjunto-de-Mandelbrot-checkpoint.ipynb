```julia
using PyPlot
```

```julia
# julia set
# (the familiar mandelbrot set is obtained by setting c==z initially)
function mandelbrot(z,c; maxiter=300)
    for n = 1:maxiter
        if abs2(z) > 4
            return n-1
        end
        z = z*z + c
    end
    return maxiter
end
```

```julia
@time m = [ mandelbrot(complex(r,i),complex(r,i)) for i=1:-.002:-1, r=-1.5:.002:1.5 ];
```

```julia
get_cmap("RdGy")
```

```julia
imshow(m, cmap="RdGy", extent=[-1.5,1.5,-1,1])
```
